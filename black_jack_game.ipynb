```python
suits = ('Hearts' , 'Diamonds' , 'Spades' , 'Clubs')
ranks = ('Two' , 'Three' , 'Four', 'Five' , 'Six' , 'Seven', 'Eight' , 'Nine' , 'Ten', 'Jack', 'Queen', 'King' , 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python

import random
```

```python
class Cards:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return(f'{self.rank} of {self.suit}')
```

```python
class Deck:
    def __init__(self):
        self.total_deck = []
        for suit in suits:
            for rank in ranks :
                self.total_deck.append(Cards(suit, rank))
    def shuffle(self):
        random.shuffle(self.total_deck)
    def deal_one(self):
        return self.total_deck.pop()
    def __len__(self):
        return len(self.total_deck)
```

```python
class Chips():
    def add_chips(self, new_chips):
            self.chips = self.chips + new_chips
            print(f'You new stand is on {self.chips} chips')
    def lose_chips(self, lost_chips):
            self.chips = self.chips - lost_chips
            print(f'You new stand is on {self.chips} chips')

   
        
```
